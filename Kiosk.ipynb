```python
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 2500, 3000, 2000, 2500, 3000]
        self.menu_list = []
        self.price_list = []
        self.inner_list = []
        self.t_temp = None
    # 메뉴 출력 메서드
    def menu_print(self):
       for i in range(len(self.menu)):
           print(i+1, self.menu[i], " : ", self.price[i], "원")

    # 주문 메서드
    def menu_select(self):
        print()
        try: 
            n = int(input("음료의 번호를 입력해주세요. \n결제 하시려면 '0'을 입력하세요. "))
            
            if n < 0 or n > len(self.menu):
                raise Exception
        except:
            print("잘못 입력 하셨습니다. 다시 입력해 주세요. ") 
            self.menu_select()
        else:
            if n != 0:
                self.temp_select()
                print(self.t_temp, self.menu[n-1], " : ", self.price[n-1], "원")
                self.inner_list.append(self.menu[n-1])
                self.inner_list.append(self.price[n-1])
                self.inner_list.append(self.t_temp)
                self.menu_list.append(self.inner_list)
                self.inner_list = []
                [print(i) for i in self.menu_list]
                    
                self.menu_select()
            else:
                self.pay()
            
    # 온도 선택
    def temp_select(self):
        try:
            t = input("음료의 온도를 선택해 주세요. 1: HOT, 2: ICE ")
            if t != '1' and t != '2':
                raise Exception
        except:
            print("잘못 입력 하셨습니다. 다시 입력해 주세요. ")
            self.temp_select()
        else:
            if t == '1':
                self.t_temp = 'HOT'
            else:
                self.t_temp = 'ICE'
    # 지불
    def pay(self):
        print()
        try:
            user = input("결제 방법을 선택해 주세요. \n 현금: 현금 or 1 \n 카드: 카드 or 2 ")
            if user == '현금' or user == '1':
                print("직원을 호출 하겠습니다. ")
            elif user == '카드' or user == '2':
                print("카드를 알맞은 방향으로 꽂아 주세요. ")
            else:
                raise Exception
        except:
            print("잘못 입력하셨습니다. ")
            self.pay
        

    # 주문서 출력 
    def table(self):
        print()
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        for i in range(len(self.menu_list)):
            print(self.menu_list[i])
            
        [self.price_list.append(self.menu_list[i][1]) for i in range(len(self.menu_list))]

        print('합계 금액 :', sum(self.price_list))

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')
        

        
a = Kiosk()
a.menu_print()
a.menu_select()
a.table()
```

```python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print(f'직사각형 함수는 {self.rect()}입니다.')

        elif self.square == 2:
            print(f'평행사변형 함수는 {self.par()}입니다.')
        
        elif self.square == 3:
            print(f'사다리꼴 함수는 {self.trape()}입니다.')
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        width, vertical = map(int, input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>>').split(','))
        area = width * vertical
        result = '평행사변형의 넓이는 : ' + str(area)
        return result

    def trape(self):
        lWidth, hWidth, vertical = map(int, input('밑변, 윗변, 높이를 입력하세요. 예시 : 밑변,윗변,높이\n >>>').split(','))
        area = ((lWidth + hWidth) * vertical)/2
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result

a = Square()
```
